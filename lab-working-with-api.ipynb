```python
#normalise the response

data = response.json()
formatted_data = json.dumps(data, indent=4)
formatted_data
```

```python
# Check if the request was successful
if response.status_code == 200:
    data = response.json()

    # Find the trading pair with the highest trading volume
    pairs = data
    highest_volume_pair = max(pairs, key=lambda x: float(x["quoteVolume"])) # we need to extract the max quoteVolume of each cion and convert it to a float
    symbol = highest_volume_pair["symbol"]
    trading_volume = highest_volume_pair["quoteVolume"]

    print("The trading pair with the highest trading volume in the last 24 hours is {} with a volume of {}.".format(symbol, trading_volume))
else:
    print("There was an error retrieving the data")
```
